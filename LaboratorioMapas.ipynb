```python
from folium.plugins import minimap
from branca.element import Figure

popuptext = '<b>CASA</b>'
my_map = folium.Map(location=[10.02074884419554, -84.1526482742486], zoom_start=16, titles='stamenterrain')

folium.Marker(location=[10.02074884419554, -84.1526482742486], popup=popuptext, icon=folium.Icon(color="orange", icon="ok-sign")).add_to(my_map)
folium.Circle(location=[10.02074884419554, -84.1526482742486], color="purple", weight=6, fill_color="red", fill_opacity=0.5, tooltin="CASA").add_to(my_map)

minimap = MiniMap(title_layer='stamenterrain', position='bottomright')
my_map.add_child(minimap)

my_map.save('C:mapa.html')
my_map
```
