```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/Shooting-and-Firearm-Discharges-Open-Data.csv')

data=pd.DataFrame(df)

dataResultado=data.loc[(data['MONTH']>='September')]

print(dataResultado)
```

```python

ciudad_mas_muertes=df.groupby("NEIGHBOURHOOD_158")["DEATH"].mean()
ciudad_mas_muertes.head(10).plot.barh(color="gray")
plt.xlabel('Muertes')
plt.ylabel('Barrios')
plt.title('Promedio muertes por Barrio')
plt.show()
```
